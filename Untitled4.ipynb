## Step 3: Perform Optical Character Recognition (OCR)

Now, we'll use `easyocr` to detect and recognize text within the uploaded image. The code will also draw bounding boxes around the detected text, which can be thought of as the 'highlighted' editable regions.

```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Initialize EasyOCR reader (this might take some time on the first run to download models)
# You can specify languages, e.g., ['en'] for English
reader = easyocr.Reader(['en'])

# Read the image using OpenCV
image_np = np.array(bytearray(uploaded[image_filename]), dtype=np.uint8)
img = cv2.imdecode(image_np, cv2.IMREAD_COLOR)

# Perform OCR
results = reader.readtext(img)

# Create a copy of the image to draw bounding boxes on
img_with_boxes = img.copy()

# Draw bounding boxes and print detected text
detected_texts = []
for (bbox, text, prob) in results:
    # Get coordinates
    (top_left, top_right, bottom_right, bottom_left) = bbox
    top_left = tuple(map(int, top_left))
    bottom_right = tuple(map(int, bottom_right))

    # Draw rectangle on image
    cv2.rectangle(img_with_boxes, top_left, bottom_right, (0, 255, 0), 2) # Green box

    # Put text near the box (optional, for debugging/display)
    cv2.putText(img_with_boxes, text, (top_left[0], top_left[1] - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2) # Red text

    detected_texts.append({'bbox': bbox, 'text': text, 'probability': prob})

# Display the image with detected text
plt.figure(figsize=(15, 15))
plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title('Image with Detected Text')
plt.show()

print("\nDetected Texts:")
for i, item in enumerate(detected_texts):
    print(f"  {i+1}. Text: '{item['text']}' (Confidence: {item['probability']:.2f})")

# Store results for potential further processing
# You would typically use these 'results' to map cursor position to text regions
# for interactive editing. This part would require a custom frontend application.
image_ocr_results = {'original_image': img, 'detected_texts': detected_texts}
```

## Step 4: Edit and Replace Text

Now, let's select a detected text, remove it, and replace it with new content. Follow the prompts below to choose which text to edit and what to replace it with.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Ensure image_ocr_results is available from the previous step
if 'image_ocr_results' not in locals():
    print("Error: OCR results not found. Please run the previous cells first.")
else:
    original_img = image_ocr_results['original_image'].copy()
    detected_texts = image_ocr_results['detected_texts']

    print("\nAvailable texts for editing:")
    for i, item in enumerate(detected_texts):
        print(f"  {i+1}. Text: '{item['text']}'")

    try:
        text_index_to_edit = int(input("Enter the number of the text you want to edit: ")) - 1
        if not (0 <= text_index_to_edit < len(detected_texts)):
            raise ValueError("Invalid index.")

        new_text_content = input("Enter the new text content: ")

        selected_text_info = detected_texts[text_index_to_edit]
        bbox = selected_text_info['bbox']

        # Convert bounding box to integer coordinates for OpenCV
        (top_left, top_right, bottom_right, bottom_left) = bbox
        x_min, y_min = int(min(top_left[0], bottom_left[0], top_right[0], bottom_right[0])), int(min(top_left[1], bottom_left[1], top_right[1], bottom_right[1]))
        x_max, y_max = int(max(top_left[0], bottom_left[0], top_right[0], bottom_right[0])), int(max(top_left[1], bottom_left[1], top_right[1], bottom_right[1]))

        # Add a small buffer to the bounding box for better inpainting
        buffer = 5
        x_min = max(0, x_min - buffer)
        y_min = max(0, y_min - buffer)
        x_max = min(original_img.shape[1], x_max + buffer)
        y_max = min(original_img.shape[0], y_max + buffer)

        # Create a mask for inpainting
        mask = np.zeros(original_img.shape[:2], dtype="uint8")
        cv2.rectangle(mask, (x_min, y_min), (x_max, y_max), 255, -1)

        # Perform inpainting to remove the old text
        edited_img = cv2.inpaint(original_img, mask, 3, cv2.INPAINT_TELEA)

        # Add new text (simplistic, does not match original font/style)
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1.0 # Adjust as needed
        font_thickness = 2 # Adjust as needed
        color = (255, 255, 255) # White color for new text

        # Get text size to center it or fit within the bbox
        (text_width, text_height), baseline = cv2.getTextSize(new_text_content, font, font_scale, font_thickness)

        # Calculate position to draw new text (e.g., centered within the bounding box)
        text_x = x_min + (x_max - x_min - text_width) // 2
        text_y = y_min + (y_max - y_min + text_height) // 2

        # Ensure text does not go out of bounds
        text_x = max(0, text_x)
        text_y = max(text_height, text_y) # Ensure text is not cut off at the top

        cv2.putText(edited_img, new_text_content, (text_x, text_y), font, font_scale, color, font_thickness, cv2.LINE_AA)

        # Display the image with the replaced text
        plt.figure(figsize=(15, 15))
        plt.imshow(cv2.cvtColor(edited_img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f'Image with Text Edited: "{selected_text_info['text']}" -> "{new_text_content}"')
        plt.show()

        print(f"Successfully replaced '{selected_text_info['text']}' with '{new_text_content}'.")

    except ValueError as e:
        print(f"Invalid input: {e}")
    except IndexError:
        print("Selected text number is out of range.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

## Interactive Text Editor Dashboard

This section provides an interactive dashboard to upload an image, perform OCR, and edit detected text. Run the cell below to activate the interface.

**Note**: The first time you run this, EasyOCR might download models, which can take a few minutes.

```python
import ipywidgets as widgets
from IPython.display import display, clear_output, Image as IPImage
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab import files
import io

# --- Global variables to store state ---
current_img = None
detection_results = []
reader = None

# --- Output widgets for dynamic content ---
output_area_upload = widgets.Output()
output_area_ocr = widgets.Output()
output_area_edit = widgets.Output()
output_area_status = widgets.Output()

# --- Widget Definitions ---
upload_button = widgets.FileUpload(
    accept='image/*',
    multiple=False,
    description='Upload Image',
    button_style='info'
)

text_selector = widgets.Dropdown(
    options=[],
    description='Select Text:',
    disabled=True,
    layout=widgets.Layout(width='auto')
)

new_text_input = widgets.Text(
    value='',
    placeholder='Enter new text',
    description='New Text:',
    disabled=True,
    layout=widgets.Layout(width='auto')
)

edit_button = widgets.Button(
    description='Apply Edit',
    disabled=True,
    button_style='success'
)

# --- Functions ---
def initialize_ocr_reader():
    global reader
    if reader is None:
        with output_area_status:
            print("Initializing EasyOCR reader (downloading models if necessary)... This may take a moment.")
        reader = easyocr.Reader(['en'])
        with output_area_status:
            clear_output(wait=True)
            print("EasyOCR reader initialized.")

def perform_ocr_and_display(image_data):
    global current_img, detection_results
    with output_area_ocr:
        clear_output(wait=True)
        if image_data is None:
            print("No image uploaded.")
            return

        initialize_ocr_reader()

        # Read the image using OpenCV
        image_np = np.array(bytearray(image_data), dtype=np.uint8)
        current_img = cv2.imdecode(image_np, cv2.IMREAD_COLOR)

        if current_img is None:
            print("Error: Could not decode image. Please upload a valid image file.")
            return

        print("Performing OCR...")
        results = reader.readtext(current_img)
        detection_results = []

        img_with_boxes = current_img.copy()
        text_options = []

        if not results:
            print("No text detected in the image.")
            text_selector.options = []
            text_selector.disabled = True
            new_text_input.disabled = True
            edit_button.disabled = True
            plt.figure(figsize=(10, 10))
            plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
            plt.axis('off')
            plt.title('Original Image (No text detected)')
            plt.show()
            return

        for i, (bbox, text, prob) in enumerate(results):
            detection_results.append({'bbox': bbox, 'text': text, 'probability': prob})
            text_options.append((f"{i+1}. {text} (Confidence: {prob:.2f})", i))

            # Draw rectangle on image for display
            (top_left, top_right, bottom_right, bottom_left) = bbox
            top_left_int = tuple(map(int, top_left))
            bottom_right_int = tuple(map(int, bottom_right))
            cv2.rectangle(img_with_boxes, top_left_int, bottom_right_int, (0, 255, 0), 2) # Green box
            cv2.putText(img_with_boxes, str(i+1), (top_left_int[0] + 5, top_left_int[1] + 15),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2) # Red index for selection

        text_selector.options = text_options
        text_selector.disabled = False
        new_text_input.disabled = False
        edit_button.disabled = False

        plt.figure(figsize=(10, 10))
        plt.imshow(cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title('Detected Text Regions (indexed for selection)')
        plt.show()
        print("OCR complete. Select text to edit below.")

def apply_text_edit(text_index, new_content):
    global current_img
    with output_area_edit:
        clear_output(wait=True)
        if current_img is None or not detection_results:
            print("No image or detected text to edit.")
            return

        if not (0 <= text_index < len(detection_results)):
            print("Invalid text selection.")
            return

        selected_text_info = detection_results[text_index]
        bbox = selected_text_info['bbox']

        # Convert bounding box to integer coordinates for OpenCV
        (top_left, top_right, bottom_right, bottom_left) = bbox
        x_min, y_min = int(min(top_left[0], bottom_left[0], top_right[0], bottom_right[0])), int(min(top_left[1], bottom_left[1], top_right[1], bottom_right[1]))
        x_max, y_max = int(max(top_left[0], bottom_left[0], top_right[0], bottom_right[0])), int(max(top_left[1], bottom_left[1], top_right[1], bottom_right[1]))

        # Add a small buffer to the bounding box for better inpainting
        buffer = 5
        x_min = max(0, x_min - buffer)
        y_min = max(0, y_min - buffer)
        x_max = min(current_img.shape[1], x_max + buffer)
        y_max = min(current_img.shape[0], y_max + buffer)

        # Create a mask for inpainting
        mask = np.zeros(current_img.shape[:2], dtype="uint8")
        cv2.rectangle(mask, (x_min, y_min), (x_max, y_max), 255, -1)

        # Perform inpainting to remove the old text
        edited_img_temp = cv2.inpaint(current_img, mask, 3, cv2.INPAINT_TELEA)

        # Add new text (simplistic, does not match original font/style)
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1.0 # Default font scale
        font_thickness = 2 # Default font thickness
        color = (255, 255, 255) # White color for new text

        # Get text size to center it or fit within the bbox dynamically
        (text_width, text_height), baseline = cv2.getTextSize(new_content, font, font_scale, font_thickness)

        # Adjust font_scale to fit within the bounding box if necessary
        if text_width > (x_max - x_min) or text_height > (y_max - y_min):
            scale_w = (x_max - x_min) / text_width
            scale_h = (y_max - y_min) / text_height
            font_scale = min(scale_w, scale_h) * 0.9 # Use 90% of max scale to give some padding
            (text_width, text_height), baseline = cv2.getTextSize(new_content, font, font_scale, font_thickness)

        text_x = x_min + (x_max - x_min - text_width) // 2
        text_y = y_min + (y_max - y_min + text_height) // 2

        # Ensure text does not go out of bounds
        text_x = max(0, text_x)
        text_y = max(text_height, text_y) # Ensure text is not cut off at the top

        cv2.putText(edited_img_temp, new_content, (text_x, text_y), font, font_scale, color, font_thickness, cv2.LINE_AA)

        current_img = edited_img_temp # Update the current image state

        plt.figure(figsize=(10, 10))
        plt.imshow(cv2.cvtColor(current_img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.title(f"Image with Edited Text: '{selected_text_info['text']}' -> '{new_content}'")
        plt.show()
        print(f"Successfully replaced '{selected_text_info['text']}' with '{new_content}'.")

# --- Event Handlers ---
def on_upload_change(change):
    with output_area_upload:
        clear_output(wait=True)
        if upload_button.value: # Check if the dictionary is not empty
            # FileUpload.value is a dict: {'filename': {'content': b'...', 'metadata': {...}}}
            # We need to get the actual file content and name from the first (and only) item
            uploaded_file_data = list(upload_button.value.values())[0]
            image_content = uploaded_file_data['content']
            file_name = uploaded_file_data['metadata']['name']

            print(f"Uploaded: {file_name}")
            display(IPImage(image_content))
            perform_ocr_and_display(image_content)
        else:
            print("No file selected or upload cancelled.")

def on_edit_button_click(b):
    if text_selector.value is not None and new_text_input.value:
        apply_text_edit(text_selector.value, new_text_input.value)
    else:
        with output_area_edit:
            print("Please select text and enter new content.")

# --- Register Callbacks ---
upload_button.observe(on_upload_change, names='value')
edit_button.on_click(on_edit_button_click)

# --- Layout and Display ---
display(output_area_status,
        widgets.VBox([
            widgets.HBox([upload_button]),
            output_area_upload,
            output_area_ocr,
            widgets.HBox([text_selector, new_text_input, edit_button]),
            output_area_edit
        ])
       )

# Initial setup to make sure OCR reader is ready for subsequent operations
initialize_ocr_reader()
```
